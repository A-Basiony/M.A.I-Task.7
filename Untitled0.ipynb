```python
import numpy as np
import matplotlib.pyplot as plt

# Given signal description
max_voltage = 1.4  # Peak voltage in volts
duration = 2       # Duration in seconds

# Time points for the original signal (continuous)
t = np.linspace(0, duration, 1000)

# Doom signal function approximation (assumed parabolic based on image)
def doom_signal(t):
    return 4 * max_voltage / duration**2 * t * (duration - t)

# Original continuous signal
continuous_signal = doom_signal(t)

# Sampling points
sampling_points_025 = np.arange(0, duration + 0.25, 0.25)
sampling_points_05 = np.arange(0, duration + 0.5, 0.5)
sampling_points_1 = np.arange(0, duration + 1, 1)

# Sampled signals
sampled_signal_025 = doom_signal(sampling_points_025)
sampled_signal_05 = doom_signal(sampling_points_05)
sampled_signal_1 = doom_signal(sampling_points_1)

# Quantization levels
# 3-bit (8 levels)
quant_levels_3bit = 8
q_levels_3bit = np.linspace(0, max_voltage, quant_levels_3bit)

# 2-bit (4 levels)
quant_levels_2bit = 4
q_levels_2bit = np.linspace(0, max_voltage, quant_levels_2bit)

# Quantization function
def quantize(signal, levels):
    quantized_signal = np.zeros_like(signal)
    for i, value in enumerate(signal):
        idx = (np.abs(levels - value)).argmin()
        quantized_signal[i] = levels[idx]
    return quantized_signal

# Quantized signals
quantized_3bit_025 = quantize(sampled_signal_025, q_levels_3bit)
quantized_3bit_05 = quantize(sampled_signal_05, q_levels_3bit)
quantized_3bit_1 = quantize(sampled_signal_1, q_levels_3bit)
quantized_2bit_025 = quantize(sampled_signal_025, q_levels_2bit)

# Plotting
fig, axs = plt.subplots(4, 1, figsize=(10, 16))

# Plot 3-bit Encoder with Sampling Time = 0.25s
axs[0].plot(t, continuous_signal, label='Analog Signal')
axs[0].stem(sampling_points_025, quantized_3bit_025, 'r', basefmt=" ", label='Quantized Signal (3-bit, 0.25s)')
axs[0].set_title('3-bit Encoder, Sampling Time = 0.25s')
axs[0].legend()

# Plot 3-bit Encoder with Sampling Time = 0.5s
axs[1].plot(t, continuous_signal, label='Analog Signal')
axs[1].stem(sampling_points_05, quantized_3bit_05, 'r', basefmt=" ", label='Quantized Signal (3-bit, 0.5s)')
axs[1].set_title('3-bit Encoder, Sampling Time = 0.5s')
axs[1].legend()

# Plot 3-bit Encoder with Sampling Time = 1s
axs[2].plot(t, continuous_signal, label='Analog Signal')
axs[2].stem(sampling_points_1, quantized_3bit_1, 'r', basefmt=" ", label='Quantized Signal (3-bit, 1s)')
axs[2].set_title('3-bit Encoder, Sampling Time = 1s')
axs[2].legend()

# Plot 2-bit Encoder with Sampling Time = 0.25s
axs[3].plot(t, continuous_signal, label='Analog Signal')
axs[3].stem(sampling_points_025, quantized_2bit_025, 'r', basefmt=" ", label='Quantized Signal (2-bit, 0.25s)')
axs[3].set_title('2-bit Encoder, Sampling Time = 0.25s')
axs[3].legend()

plt.tight_layout()
plt.show()
```
